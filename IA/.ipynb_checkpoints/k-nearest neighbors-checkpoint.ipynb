```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import KFold
```

```python
iris = pd.read_csv('/home/fernando-dlz/Documents/artificial_inteligence/iris.data', header=None)
```

```python
def clasificacion(base_datos):
    name = base_datos.iloc[:, -1]

    #FOR PARA ESTABLECER LOS INDICES DE CADA CLASE
    i = 0
    index_list = []
    indexes = []
    for j in range(base_datos.shape[0]):
        if name[j] != name[i]:
            indexes.append(i)
            indexes.append(j)
            index_list.append(indexes)
            i = j
            indexes = []
    indexes.append(i)
    indexes.append(j+1)
    index_list.append(indexes)    
        
    #FOR PARA GUARDAR LOS DATOS DE UNA CLASE COMO ELEMENTOS (TIPO PANDAS) DE UNA LISTA
    dataset = []
    for e in index_list:
        dataset.append(base_datos.iloc[e[0]:e[1], :])
    
    return dataset
```

```python
def distancia(punto1, punto2):    
    dimensions = punto2.shape[0] - 1
    
    #CONVERTIMOS LOS OBJETOS DE PANDAS EN LISTAS CRUDAS
    item1 = []
    l = 0
    for x in range(dimensions):
        item1.append(punto1[l])
        l += 1

    item2 = []
    l = 0
    for x in range(dimensions):
        item2.append(punto2[l])
        l += 1
        
    suma = 0
    i = 0
    #FOR PARA CALCULAR DISTANCIA
    for e in item1:
        suma += (e-item2[i])**2
        i += 1

    return math.sqrt(suma)
```

```python
def translate(data_base, index):
    return data_base[4][index]
```

```python
def knn(punto, dataset, k=3):
    names_list = []
    indexes_list = []
    
    #LLENAMOS EL INDEX_LIST CON -1
    for i in range(k):
        indexes_list.append(-1)

    #ENCONTRAMOS EL VALOR MAXIMO
    maximo = 0
    for clase in dataset:
        for e in range(clase.shape[0]):
            if distancia(punto, clase.iloc[e]) > maximo:
                maximo = distancia(punto, clase.iloc[e])

    #FOR PARA LLENAR INDEXES_LIST CON LOS INDICES DE LOS NUMEROS CON LAS DISTANCIAS MAS PEQUEÃ‘AS
    minimo = maximo
    for i in range(k):
        for clase in dataset:
            for e in range(clase.shape[0]):
                if clase.iloc[e].name not in indexes_list:
                    if distancia(punto, clase.iloc[e]) != 0:
                        if distancia(punto, clase.iloc[e]) < minimo:
                            minimo = distancia(punto, clase.iloc[e])
                            indexes_list[i] = clase.iloc[e].name
        minimo = maximo

    #FOR PARA CONVERTIR LOS INDICES EN SUS RESPECTIVAS CLASES
    for e in indexes_list:
        names_list.append(translate(iris, e))

    return max(names_list)
```

```python
def kfolds(dataset, k = 10):
    kf = KFold(n_splits=k)

    #FOR QUE GENERA LA ESTRUCTURA DE MI LISTA
    index_kfolds = []
    for i in range(k):
        index_kfolds.append([])

    #FOR QUE VA LLENANDO MI LISTA CON CADA FOLD
    i = 0
    for clase in dataset:
        for training_index, test_index in kf.split(clase):
            item = []
            item.append(test_index)
            item.append(training_index)
            index_kfolds[i].append(item)
            i +=1
        i = 0
        
    return index_kfolds
```

```python
def get_training(test_index, dataset):
    margin = dataset.iloc[0].name
    
    for e in test_index:
        dataset = dataset.drop(e+margin)
    
    return dataset.iloc[:, :]
```

```python
def convert_point(index, clase):
    dimensions = clase.shape[1] -1
    margin = clase.iloc[0].name

    item = []
    for i in range(dimensions):
        item.append(clase[i][index+margin])
    
    return item
```

```python
#Cross Validation
dataset = clasificacion(iris)
folds = kfolds(dataset)
n = 1

for fold in folds:
    i = 0
    
    # FOR PARA OBTENER UN PANDAS ARRAY CON LOS DATOS DE TRAINING
    pandas_dataset = []
    for clase in fold:
        pandas_dataset.append(get_training(clase[0], dataset[i]))
        i += 1
    
    print('#Test', n)
    n += 1
    
    #FOR PARA SACAR LAS DISTANCIAS
    i = 0
    accuracy_test = []
    accuracy_train = []

    for clase in fold:
        right_test = 0
        total_test = 0
        right_train = 0
        total_train = 0
        
        #FOR PARA EVALUAR EL TEST_SET
        for point_index in clase[0]:
            if knn(convert_point(point_index, dataset[i]), pandas_dataset) == dataset[i].iloc[point_index,-1]:
                right_test +=1
            total_test += 1
        accuracy_test.append(right_test/total_test)
                
        #FOR PARA EVALUAR EL TRAIN_SET
        for point_index in clase[1]:
            if knn(convert_point(point_index, dataset[i]), pandas_dataset) == dataset[i].iloc[point_index,-1]:
                right_train += 1
            total_train += 1
        accuracy_train.append(right_train/total_train)
        
        i += 1
        
    #CALCULO MEDIAS
    media_test = sum(accuracy_test)/len(accuracy_test)
    media_train = sum(accuracy_train)/len(accuracy_train)
    
    #CALCULO DESVIACION
    desviacion_test = 0
    for e in accuracy_test:
        desviacion_test += (e-media_test)**2
    desviacion_test = math.sqrt(desviacion_test/len(accuracy_test))

    desviacion_train = 0
    for e in accuracy_train:
        desviacion_train += (e-media_train)**2
    desviacion_train = math.sqrt(desviacion_train/len(accuracy_train))

    print('\t\tTest \t\t Train\nmedia:\t\t{:5.4f}\t\t{:5.4f}\ndesviacion:\t{:5.5f}\t\t{:5.4f}'.format(media_test, media_train, desviacion_test, desviacion_train))
    print()
```
