```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
iris = pd.read_csv('iris.data', header=None)
```

```python
def clasificacion(base_datos):
    name = base_datos.iloc[:, -1]
    count = 0
    
    k = 0
    while(k < 100):
        try:
            name[k]
            break
        except:
            k += 1
    
    i = 0
    while(i < 200):
        try:
            name[i]
            if name[i] != name[k]:
                break
            count += 1
            i += 1
        except:
            i += 1
            
    dataset = []
    num_clases = base_datos.shape[0]/count
    dimensiones = base_datos.shape[1] - 1
    dimensiones_list = []
    for x in range(dimensiones):
        dimensiones_list.append(x)

    for i in range(int(num_clases)):
        dataset.append(base_datos.iloc[i*count:i*count+count, dimensiones_list].values)
    
    return dataset
```

```python
def media(clase):
    sum = np.sum(clase, axis=0)
    dimensiones = clase.shape[0]
    media_lista = np.array([])
        
    for i in sum:
        media_lista = np.append(media_lista, i/dimensiones)
        
    return media_lista
```

```python
def varianza(clase):
    
    dimensiones = clase.shape[1]
    longitud = clase.shape[0]
    med = media(clase)
    

    varianza_lista = []
    
    for i in range(dimensiones):
        varianza_suma = 0
        for j in range(longitud):
            varianza_suma += (clase[j][i] - med[i])**2
        varianza_lista.append(varianza_suma/(longitud-1))
            
    return varianza_lista
```

```python
def probabilidad(punto, db):
    dataset = clasificacion(db)
    
    productos = []
    i = 0
    for i in dataset:
        med = media(i)
        vari = varianza(i)
        prob = 1
        for j in range(int(dataset[0].shape[1])):
            prob *= (math.e**((-(punto[j] - med[j])**2)/(2*vari[j])))/math.sqrt(2*math.pi*vari[j])*(1/len(dataset)) 
        productos.append(prob)
    
    probabilidades = []
    for i in range(len(productos)):
        probabilidades.append(productos[i]/sum(productos))
    
    prob = probabilidades
    
    name = ''
    if (prob[0] > prob[1] and prob[0] > prob[2]):
        name = 'setosa'
    elif (prob[1] > prob[0] and prob[1] > prob[2]):
        name = 'versicolor'
    elif (prob[2] > prob[0] and prob[2] > prob[1]):
        name = 'virginia'

    print(name)
    
    return name
```

```python
def folds(data_base, ki, k=10):

    name = data_base.iloc[:, -1]
    count = 0
    while (name[count] == name[0]):
        count += 1

    size_folds = count/k
    num_clases = data_base.shape[0]/count   
    dimensiones = data_base.shape[1] - 1
    
    testing_list = []
    for l in range(int(num_clases)):
        items = []
        for m in range(int(size_folds)):
                item = []
                for x in range(dimensiones):
                    item.append(data_base[x][int(size_folds)*ki+l*count+m])
                items.append(item)
                data_base = data_base.drop(int(size_folds)*ki+l*count+m)
        testing_list.append(items)
        
    return testing_list, data_base
```

```python
#Testing

t = 1
for i in range(10):
    print('#Test', t)
    testing_list, data_set = folds(iris, i)
    for testing in testing_list:
        for k in testing:
            probabilidad(k, data_set)
        print('')
    t += 1
    
```
